```python
from pathlib import Pathlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from fastapi import FastAPI
from fastapi.responses import FileResponse
import io

app = FastAPI()



data_path=Path.cwd().parent/"resource/prompts.csv"

def analyze_prompts():
    # 1. Load data using Pandas
    df = pd.read_csv(data_path)

    # 2. Count tokens using string operations (split by whitespace)
    df['token_count'] = df['prompt'].apply(lambda x: len(str(x).split()))

    # 3. Calculate stats using NumPy
    stats = {
        "average": float(np.mean(df['token_count'])),
        "min": int(np.min(df['token_count'])),
        "max": int(np.max(df['token_count'])),
        "total_prompts": len(df)
    }

    # 4. Create Bar Chart using Matplotlib
    plt.figure(figsize=(8, 4))
    plt.bar(df['id'].astype(str), df['token_count'], color='skyblue')
    plt.xlabel('Prompt ID')
    plt.ylabel('Token Count')
    plt.title('Prompt Length Analysis')
    
    # Save plot to a buffer or file
    plot_path = "prompt_analysis.png"
    plt.savefig(plot_path)
    plt.close()
    
    return stats, plot_path

@app.get("/analyze")
async def get_analysis():
    stats, plot_path = analyze_prompts()
    return {
        "statistics": stats,
        "chart_url": "/chart"
    }

@app.get("/chart")
async def get_chart():
    return FileResponse("prompt_analysis.png")

if __name__ == "__main__":
    import uvicorn
    uvicorn.run(app, host="0.0.0.0", port=8000)
```
